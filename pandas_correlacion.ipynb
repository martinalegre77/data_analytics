#  Correlación
## Magnitud y direccion de una relación lineal entre 2 variables

```python
# importamos librerias
import matplotlib.pyplot as plt
import pandas as pd

personas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/csv/personas.csv')
```

```python
personas.head()
```

## Relación: Altura y Peso

```python
plt.scatter(personas['altura'], personas['peso'])
plt.xlabel('Altura (cms)')
plt.ylabel('Peso (Kgs)')
plt.show()
```

### Hay una cierta relación entre las variables ya que se percibe una pendiente entre el aumento de la altura y el aumento del peso (relación fuerte).

## Relación: Ingreso y Horas Trabajadas

```python
plt.scatter(personas['ingreso'], personas['horas_trabajadas'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Horas Trabajadas')
plt.show()
```

### Aquí se percibe mayor dispersión en los datos por lo que se puede deducir que la relación no es tan directa. Traducido, trabajar más horas no implica que se obtendrán mayores ingresos.

## Relación: Ingreso y Ausencias

```python
plt.scatter(personas['ingreso'], personas['ausencias'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Ausencias')
plt.show()
```

### Esta relación tiene una magnitud muy fuerte (las personas con mayores ausencias tienen menores ingresos). La pendiente es negativa, la relación es negativa.

## Relación: Ingreso y Peso de una persona

```python
plt.scatter(personas['ingreso'], personas['peso'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Peso (Kgs)')
plt.show()
```

### Datos completamente dispersos, no se pueden ajustar a una recta, no hay una relación lineal.

# Matriz de Correlación

```python
personas.corr()
# coeficientes de correlación
```

### -1 (existe una relación fuerte negativa) o 1 (existe una relación fuerte positiva). Valor 0 la relación es debil.
- De 0.3 a 0.5 = correlación baja.
- De 0.5 a 0.7 = correlación moderada.
- De 0.7 a 1 = correlación fuerte.


# Gráfica de la matriz de correlación

```python
matriz = personas.corr()
plt.matshow(matriz, cmap='bwr', vmin=-1, vmax=1) # b=blue, w=white, r=red
plt.xticks(range(5), personas.columns, rotation=90) # agregamos los nombres a las columnas
plt.yticks(range(5), personas.columns)

for i in range(len(matriz.columns)):
  for j in range(len(matriz.columns)):
    plt.text(i, j, round(matriz.iloc[i, j], 2), ha='center', va='center')

plt.colorbar() # agrega la barra de color
plt.show()
```
