# Predecir el ingreso de las personas en función de las horas semanales trabajadas.

## Creación del Modelo de Regresión Lineal Simple

```python
from sklearn import linear_model

# se crea un creador de modelos de regresion lineal
regresion = linear_model.LinearRegression()
# se preparan los datos
horas = datos['horas'].values.reshape((-1, 1)) # los datos deben estar en un vector
# se crea el modelo
modelo = regresion.fit(horas, datos['ingreso'])

print('Intersección (b)', modelo.intercept_)
print('Pendiente (m)', modelo.coef_)

# Predicciones
entrada = [[39], [40], [43], [44]]
modelo.predict(entrada)

# Gráfica
plt.scatter(entrada, modelo.predict(entrada), color='green')
plt.plot(entrada, modelo.predict(entrada), color='black') # graficamos la linea

plt.ylabel('Ingreso ($)')
plt.xlabel('Promedio de horas semanales trabajadas')
plt.scatter(datos['horas'], datos['ingreso'], color='pink')
plt.show()
```
